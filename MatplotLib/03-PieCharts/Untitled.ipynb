```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [60,40]

plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## with labels

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [60,40]
labels = ['Sixty','Forty']

plt.pie(slices, labels = labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## t have a wedge color

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [60,40]
labels = ['Sixty','Forty']

plt.pie(slices, labels = labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## to have a differentcustom  color for different pie

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [60,40,77,10]
labels = ['Sixty','Forty','Seventy-Seven','Ten']
colors = ['#008fd5','#fc4f30','#e5ae37','#6d904f']

plt.pie(slices, labels = labels, colors = colors, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## explode concept for highliting a slice of pie chart

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
## if want to explode 4th elemnent of data
explode = [0,0,0,0.1,0]

plt.pie(slices, labels = labels, explode = explode, colors = colors, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## to have a shadow around the edges use shadow property to true

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
## if want to explode 4th elemnent of data
explode = [0,0,0,0.1,0]

plt.pie(slices, labels = labels, explode = explode, colors = colors, shadow = True, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## startangle

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
## if want to explode 4th elemnent of data
explode = [0,0,0,0.1,0]

plt.pie(slices, 
        labels = labels,
        explode = explode, 
        colors = colors,
        startangle=90, 
        shadow = True, 
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## to have a %age dispalyed for each slice what they occupied

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
## if want to explode 4th elemnent of data
explode = [0,0,0,0.1,0]

plt.pie(slices, 
        labels = labels,
        explode = explode, 
        colors = colors,
        startangle=90, 
        shadow = True, 
        autopct = '%1.1f%%',    ## for  %age dispalyed
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```
